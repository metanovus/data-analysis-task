````python
"""Четвёртная задача с предпочтениями"""

# В задаче сказано, что каждый использует две из трёх социальных сети
# добавим каждому по одной соцсети, чтобы он использовал две сети одновременно

preferences = {
    'Энн': {'Instagram', 'Not Facebook', 'Youtube'},
    'Джон': {'Not Instagram', 'Facebook', 'Youtube'},
    'Кейт': {'Not Instagram', 'Facebook', 'Youtube'},
    'Том': {'Instagram', 'Facebook', 'Not Youtube'}
    }

# Для сравнению пар, которые уже проверяли
checked_pairs = set()

for name1 in preferences:
    for name2 in preferences:
        if name1 != name2:
            # Используем frozenset для добавления в множество checked_pairs
            pair = frozenset({name1, name2})
            if pair not in checked_pairs:
                # Проверяем, есть ли общие предпочтения
                common_preferences = preferences[name1] & preferences[name2]
                checked_pairs.add(pair)
                if common_preferences:
                    print(f"У {name1} и {name2} совпадения: {', '.join(common_preferences)}")
````

Итого: если брать в рассмотрение два предпочтения, то у Джона и Кейт предпочтения совпадают по части Youtube и Facebook, а так же они оба не сидят в Instagram.

````python
"""Шестая - транспортная задача"""

# Запишем данные задачи: дистанция, скорость авто в обоих направлениях
S = 260
v1 = 80
v2 = 100

# Найдём время туда и время обратно
t1 = S / v1
t2 = S / v2

# Найдём разницу во времени в часах и переведём в минуты
difference = (t1 - t2) * 60
print(f'{difference:.0f} минут - на столько был быстрее путь назад')
````

## 2️⃣ Ответы на вопросы к ТЗ

Так как ответы на вопросы предполагают отсеивание датасета по месяцу, заранее заготовим отдельную функцию, к которой будем обращаться.

````python
def separate_df_by_month(df: pd.DataFrame, month: int) -> pd.DataFrame:
    """Отсеивает DataFrame по месяцам.

    Args:
        df: DataFrame для сортировки.
        month: Номер месяца в виде цифры (от 1 до 10).

    Returns:
        DataFrame, отсортированный по месяцам.

    Raises:
        AssertionError: Если month не находится в диапазоне от 1 до 10.
        IndexError: Если в DataFrame не найдены строки с заданным месяцем.

    """
    assert month in range(1, 11), 'Необходимо ввести число месяца (integer): от 1 до 10'

    month_names = {
        1: 'Январь', 2: 'Февраль', 3: 'Март', 4: 'Апрель', 5: 'Май',
        6: 'Июнь', 7: 'Июль', 8: 'Август', 9: 'Сентябрь', 10: 'Октябрь'
    }

    if month == 10:
        lower_index = df[df['status'] == f'{month_names[month]} 2021'].index[0]
        df_sorted = df.iloc[lower_index + 1:, :]
        return df_sorted

    try:
        lower_index = df[df['status'] == f'{month_names[month]} 2021'].index[0]
        upper_index = df[df['status'] == f'{month_names[month + 1]} 2021'].index[0]
    except IndexError:
        raise IndexError(f'Не найдены строки с месяцем {month_names[month]} 2021 или {month_names[month + 1]} 2021')

    df_sorted = df.iloc[lower_index + 1:upper_index, :]
    return df_sorted
````

### 2️⃣ Как изменялась выручка компании за рассматриваемый период? Проиллюстрируйте графиком.

Для визуализации данных будем использовать библиотеку Matplotlib. Для начала отсеим из нашего датасета только те значения, которые имеют статус `ОПЛАЧЕНО`. После чего построим несколько графиков (мини-дашборд) для более широкого анализа продаж.

Напишем всё в виде одной функции, чтобы переменные не хранились в оперативной памяти компьютера.

````python
def plot_images(df: pd.DataFrame) -> None:
    """
    Строит 6 графиков на одном выводе для анализа выручки с статусом "ОПЛАЧЕНО".

    Args:
        df: DataFrame с данными о заказах.

    Returns:
        None. Построение графиков происходит в методе plt.show().
    """
    df_paid = df[(df['status'] == 'ОПЛАЧЕНО')]
    df_grouped = df_paid.groupby('receiving_date')['sum'].sum()

    fig, axes = plt.subplots(3, 2, figsize=(16, 12))

    # 1. Линейный график
    axes[0, 0].plot(df_grouped.index, df_grouped.values, color='blue', linewidth=2)
    axes[0, 0].set_xlabel('Дата подтверждения операции')
    axes[0, 0].set_ylabel('Выручка')
    axes[0, 0].set_title('Линейный график')
    axes[0, 0].tick_params(axis='x', rotation=45, labelsize=7)

    # 2. Столбчатая диаграмма
    axes[0, 1].bar(df_grouped.index, df_grouped.values, color='skyblue')
    axes[0, 1].set_xlabel('Дата подтверждения операции')
    axes[0, 1].set_ylabel('Выручка')
    axes[0, 1].set_title('Bar chart')
    axes[0, 1].tick_params(axis='x', rotation=45)

    # 3. Гистограмма
    axes[1, 0].hist(df_paid['sum'], bins=10, edgecolor='black')
    axes[1, 0].set_xlabel('Выручка')
    axes[1, 0].set_ylabel('Количество')
    axes[1, 0].set_title('Гистограмма')

    # 4. Точечная диаграмма "Доля выручки по менеджерам в июле"
    df_sale_grouped = df_paid.groupby('sale')['sum'].sum()
    axes[1, 1].pie(df_sale_grouped, labels=df_sale_grouped.index, autopct='%1.1f%%', startangle=90)
    axes[1, 1].set_title('Доля выручки по менеджерам в июле')

    # 6. Точечная диаграмма "Доля выручки по статусу сделки в июле"
    df_new_current_grouped = df_paid.groupby('new/current')['sum'].sum()
    axes[2, 0].pie(df_new_current_grouped, labels=df_new_current_grouped.index, autopct='%1.1f%%', startangle=90)
    axes[2, 0].set_title('Доля выручки по статусу сделки в июле')

    # 7. Точечная диаграмма "Доля выручки по наличию документа в июле'"
    df_document_grouped = df_paid.groupby('document')['sum'].sum()
    axes[2, 1].pie(df_document_grouped, labels=df_document_grouped.index, autopct='%1.1f%%', startangle=90)
    axes[2, 1].set_title('Доля выручки по наличию документа в июле')

    fig.suptitle('Анализ выручки со статусом "ОПЛАЧЕНО" в июле', fontsize=16)
    plt.tight_layout()
    plt.show()
````

````python
plot_images(df_july)
````

Некоторые выводы по графикам за июль 2021 года:
* по `линейному графику` и `bar chart` видим, что максимальная выручка у нас была в начале июля (если предположить, что дата, указанная в столбце `receiving_date` коррелирует с датой фактической оплаты);

* по `гистограмме` видим, что большая часть сделок приходится на сумму до 5000 денежных единиц;
* если рассмотреть `линейный график` (где видно, что у нас падает объём выручки в июле) и соотнести с `pie chart "Доля выручки"`, где у нас указана доля выручки, то можно сделать вывод, что наша компания работает только в основном с текущими клиентами и недостаточно эффективно привлекает новых;

договора с клиентом (в рассматриваемом месяце).

Бонусы по сделкам, оригиналы для которых приходят позже рассматриваемого
месяца, считаются остатком на следующий период, который выплачивается по мере
прихода оригиналов. Необходимо вычислить остаток каждого из менеджеров на 01.07.2021.

Отсортируем наш датасет, выберем только те значения, которые датируются до 1 июля 2021 года и для удобства вычислений создадим столбец `deal_month`, чтобы опираться на месяц сделки. Также взглянем на то, какие у нас вообще менеджеры, и удалим пустые значения, если таковы имеются.

````python
# Взглянем на то, какие вообще имена у менеджеров
df['sale'].unique()
````

````python
df_balance_may = separate_df_by_month(df, 5)
df_balance_june = separate_df_by_month(df, 6)

df_balance_may['deal_month'] = pd.to_datetime('2021-05-01', format='%Y-%m-%d')
df_balance_june['deal_month'] = pd.to_datetime('2021-06-01', format='%Y-%m-%d')

df_balance = pd.concat([df_balance_may, df_balance_june])

# Удалим из датасета менеджеров с пустыми именами
df_balance = df_balance[~df_balance['sale'].isin([np.nan, '-'])]
````

Напишем функции с условиями и через метод pandas.DataFrame.apply создадим две новые колонки: `current_month_bonus` - для бонуса в текущем месяце (если месяц получения договора и месяц совершения сделки совпадают) и `next_month_bonus` - в противном случае. Так как в условии задания требуется найти остаток для менеджеров на 01.07.2021, то всё, что было выдано после июня не попадёт в новые колонки.

````python
def calculate_bonus(row: pd.Series) -> float:
    """
    Рассчитывает бонус для сделки, если оригинал договора был получен в том же месяце.

    Параметры:
    row (pd.Series): Строка из DataFrame, содержащая информацию о сделке.

    Возвращает:
    float: Бонус, рассчитанный на основе условий сделки.
    """
    # Проверка того, что оригинал получен в том же месяце
    if row['receiving_date'].month == row['deal_month'].month:
        # Проверка того, что сделка новая, статус "ОПЛАЧЕНО" и документ оригинал
        if row['new/current'] == 'новая' and row['status'] == 'ОПЛАЧЕНО':
            return row['sum'] * 0.07
        # Проверка того, что сделка текущая, статус не "ПРОСРОЧЕНО", документ оригинальный
        elif row['new/current'] == 'текущая' and row['status'] != 'ПРОСРОЧЕНО':
            if row['sum'] > 10000:
                return row['sum'] * 0.05
            else:
                return row['sum'] * 0.03
    return 0


def calculate_next_month_bonus(row: pd.Series) -> float:
    """
    Рассчитывает бонус, который должен быть перенесен на следующий месяц,
    если оригинал договора был получен позже.

    Параметры:
    row (pd.Series): Строка из DataFrame, содержащая информацию о сделке.

    Возвращает:
    float: Бонус, перенесенный на следующий месяц.
    """
    # Проверка того, что оригинал получен не в том же месяце, и этот месяц не июль
    if row['receiving_date'].month != row['deal_month'].month and row['receiving_date'].month < 7:
        # Проверка того, что сделка новая со статусом "ОПЛАЧЕНО" и был получен оригинал
        if row['new/current'] == 'новая' and row['status'] == 'ОПЛАЧЕНО':
            return row['sum'] * 0.07
        # Проверка того, что сделка текущая со статусом не "ПРОСРОЧЕНО" и был получен оригинал
        elif row['new/current'] == 'текущая' and row['status'] != 'ПРОСРОЧЕНО':
            if row['sum'] > 10000:
                return row['sum'] * 0.05
            else:
                return row['sum'] * 0.03
    return 0
````

````python
df_balance['bonus_current_month'] = df_balance.apply(calculate_bonus, axis=1)
df_balance['bonus_next_month'] = df_balance.apply(calculate_next_month_bonus, axis=1)
````

````python
df_balance_june = df_balance.groupby('sale')['bonus_current_month'].apply('sum')
df_balance_july = df_balance.groupby('sale')['bonus_next_month'].apply('sum')

df_balance_full = df_balance_june + df_balance_july
````

Теперь непосредствено выведем фамилии менеджеров и их остатки по состоянию на 01.07.2021.

````python
for manager, bonus_value in zip(df_balance_full.index, df_balance_full.values):
    print(f'{manager}: {bonus_value:.2f}')
````
